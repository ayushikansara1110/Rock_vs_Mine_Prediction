```python
# separating data according to labels
x = data.drop(columns = 60, axis = 1)
x.head()
```

```python
y = data[60]
print(y)
```

training and test data

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.1, stratify = y, random_state = 2)
# test_size gives percentage of data for testing that we need ie 0.1 gives 10% for testing
# stratify you need equal number of mines and rocks toh y ke basis pe it'll split
# random_state se bas diff splits occur
```

```python
print(x.shape, xTrain.shape, xTest.shape)
```

```python
print(y.shape, yTrain.shape, yTest.shape)
```

Model Training - Logistic Regression Model

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(xTrain, yTrain)
```

Model Evaluation

```python
# accuracy on training data
prediction1 = model.predict(xTrain)
TrainAccuracy = accuracy_score(prediction1, yTrain)
```

```python
print('Accuracy on Training Data is given as ', TrainAccuracy)
```

```python
prediction2 = model.predict(xTest)
TestAccuracy = accuracy_score(prediction2, yTest)
print('Accuracy on Test data is given as ', TestAccuracy)
```
